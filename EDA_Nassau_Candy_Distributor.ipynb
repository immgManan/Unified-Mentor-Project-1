# CREATING KPI's Groupby Products

```python
data['Product Name'].unique()             # There are 15 unique products.
```

```python
new_data = data.groupby('Product Name').agg({'Cost':'sum','Sales':'sum','Units':'sum','Gross Profit':'sum'}).reset_index()
# creating a new database which contains parameters according to product name.
```

```python
new_data.head(15)
```

Gross Margin % = Gross Profit/Sales

```python
new_data['Gross Margin %'] = ((new_data['Gross Profit']/new_data['Sales'])*100).round(2)  # creating new column
```

```python
new_data.head(15)
```

Profit Per Unit = Gross Profit/ Units

```python
new_data['Profit Per Unit'] = (new_data['Gross Profit']/new_data['Units']).round(2)  # creating new column
```

```python
new_data.head(15)
```

Revenue Contribution = Product Sales/ Total Sales

```python
Total_Sales = new_data['Sales'].sum()
print(Total_Sales)
```

```python
new_data['Revenue Contribution'] = ((new_data['Sales']/Total_Sales)*100).round(2)     # creating new column
```

```python
new_data.head(15)
```

Profit Contribution = Product Profit/Total Profit

```python
new_data['Profit contribution'] = ((new_data['Gross Profit']/(new_data['Gross Profit'].sum()))*100).round(2) # creating new column
```

```python
new_data.head(15)
```

# CREATING KPI's Groupby DIVISION

```python
data.head(1)
```

```python
data['Division'].unique()       # Three types of Division : Chocolate, Sugar and Other
```

```python
division_data = data.groupby('Division').agg({'Cost':'sum','Sales':'sum','Units':'sum','Gross Profit':'sum'}).reset_index()
```

```python
division_data.head()
```

```python
division_data['Gross Margin %'] = ((division_data['Gross Profit']/division_data['Sales'])*100).round(2) # adding gross margin %
```

```python
division_data['Profit Per Unit'] = ((division_data['Gross Profit']/division_data['Units'])*100).round(2) # adding profit per unit
```

```python
division_data['Revenue Contribution'] = (division_data['Sales']/(division_data['Sales'].sum())*100).round(2) # adding revenue contribution
```

```python
division_data['Profit contribution'] = (division_data['Gross Profit']/(division_data['Gross Profit'].sum())*100).round(2) # adding profit contribution
```
