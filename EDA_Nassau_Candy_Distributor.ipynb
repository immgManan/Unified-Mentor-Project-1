# CREATING KPI's

GROSS MARGIN % = Gross Profit/Sales

```python
data['Gross Margin %'] = ((data['Gross Profit'])/(data['Sales'])*100).round(2)           # creating column named Gross Margin %
```

```python
data.head()  # checking updated table.
```

Profit Per Unit = Gross Profit/ Units

```python
data['Profit Per Unit'] = data['Gross Profit']/data['Units']            # creating column named Profit Per Unit
```

```python
data.head()  # Checking updated table
```

Revenue Contribution = Product Sales/ Total Sales

```python
data['Product Name'].unique()             # There are 15 unique products.
```

```python
Contribution = data.groupby('Product Name')['Sales'].sum().reset_index()         # Making a new Data Frame for Revenue Contribution
```

```python
Contribution.head(15)
```

```python
Contribution['Number of Orders'] = data.groupby('Product Name')['Order ID'].count().values # creating no of orders of each candy type
```

```python
Contribution.head(15)
```

```python
Total_Sales = data['Sales'].sum()           # Calculating Total Sales
```

```python
print(Total_Sales)
```

```python
Contribution['% Revenue Contribution'] = ((Contribution['Sales']/Total_Sales)*100).round(3)
```

```python
Contribution.head(15)
```

Profit Contribution = Product Profit/ Total Profit

```python
Contribution['Profit'] = data.groupby('Product Name')['Gross Profit'].sum().values
```

```python
Contribution.head(15)
```

```python
Total_Profit = data['Gross Profit'].sum().round(2)
print(Total_Profit)
```

```python
Contribution['% Profit Contribution'] = ((Contribution['Profit']/Total_Profit)*100).round(3)  # creating new column
```
